```python
import pandas as pd

# Load the CSV
df = pd.read_csv("bank-additional-full.csv", sep=';')

# Optional: check it's loaded
df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

print("All packages loaded successfully! ðŸŽ‰")

```

```python
# Check for null values and data types
df.info()

# Check unique values in the target column
df['y'].value_counts()
```

```python
# One-hot encode categorical variables
df_encoded = pd.get_dummies(df, drop_first=True)

# Preview encoded data
df_encoded.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of target
sns.countplot(x='y', data=df)
plt.title("Term Deposit Subscription (Target Variable)")
plt.show()

# Example: Age distribution by subscription
sns.boxplot(x='y', y='age', data=df)
plt.title("Age vs. Subscription")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop('y_yes', axis=1)
y = df_encoded['y_yes']  # 1 if yes, 0 if no

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import numpy as np

feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Important Features")
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)
```

```python
# Train a new RandomForest on the resampled data
model_resampled = RandomForestClassifier(random_state=42)
model_resampled.fit(X_res, y_res)
```

```python
# Make predictions on the test set
y_pred = model_resampled.predict(X_test)

# Evaluate the model
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

```

```python
# Visual comparison of target variable before and after SMOTE
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

sns.countplot(x=y_train, ax=axs[0])
axs[0].set_title("Before SMOTE")
axs[0].set_xlabel("Term Deposit Subscription (Original)")
axs[0].set_ylabel("Count")

sns.countplot(x=y_res, ax=axs[1])
axs[1].set_title("After SMOTE")
axs[1].set_xlabel("Term Deposit Subscription (SMOTE)")
axs[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
import joblib

# Save the trained model
joblib.dump(model_resampled, 'model_resampled.pkl')

# Save feature names
joblib.dump(X.columns.tolist(), 'feature_names.pkl')

# Save test labels and predictions
joblib.dump(y_test, 'y_test.pkl')
joblib.dump(y_pred, 'y_pred.pkl')
```
